# Log File Comparison

Compares two contraction controller log files side-by-side:
- **Top row**: x, y, z, yaw (actual vs reference) for log file 1
- **Bottom row**: same for log file 2
- **RMSE**: position RMSE = √((x−x_ref)² + (y−y_ref)² + (z−z_ref)²) for both

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path

plt.style.use('seaborn-v0_8-darkgrid')
```

```python
# ===== SELECT LOG FILES HERE =====
LOG_DIR = Path('log_files/contraction_controller_px4')

FILE_1 = 'sim_contraction_figure_eight.csv'
FILE_2 = 'my_run.csv'
# =================================

print('Available log files:')
for f in sorted(LOG_DIR.glob('*.csv')):
    print(f'  {f.name}')
```

```python
def load_log(path: Path) -> pd.DataFrame:
    df = pd.read_csv(path)
    # NED -> ENU: flip z for display (positive = up)
    df['z']     = -df['z']
    df['z_ref'] = -df['z_ref']
    return df

df1 = load_log(LOG_DIR / FILE_1)
df2 = load_log(LOG_DIR / FILE_2)

for label, df, fname in [('File 1', df1, FILE_1), ('File 2', df2, FILE_2)]:
    traj = df['trajectory'].dropna().iloc[0] if 'trajectory' in df.columns else 'unknown'
    dur  = df['time'].iloc[-1] - df['time'].iloc[0]
    print(f'{label}: {fname}  |  trajectory={traj}  |  {len(df)} samples  |  {dur:.1f}s')
```

## Position & Yaw: Actual vs Reference

```python
def plot_row(axes, df, title_prefix):
    """Plot x, y, z (ENU), yaw for one log file across 4 axes."""
    coords = [
        ('x',   'x_ref',   'X (m)',       'tab:red'),
        ('y',   'y_ref',   'Y (m)',       'tab:green'),
        ('z',   'z_ref',   'Z (m, ENU)',  'tab:blue'),
        ('yaw', 'yaw_ref', 'Yaw (deg)',   'tab:purple'),
    ]
    for ax, (actual, ref, ylabel, color) in zip(axes, coords):
        if actual == 'yaw':
            a_data = np.rad2deg(df[actual])
            r_data = np.rad2deg(df[ref])
        else:
            a_data = df[actual]
            r_data = df[ref]

        ax.plot(df['time'], a_data, color=color, lw=2.0, label='Actual')
        ax.plot(df['time'], r_data, color=color, lw=1.5, ls='--', alpha=0.7, label='Reference')
        ax.set_ylabel(ylabel, fontsize=11)
        ax.legend(fontsize=9, loc='upper right')
        ax.set_title(f"{title_prefix} — {ylabel}", fontsize=10)


fig, axes = plt.subplots(2, 4, figsize=(18, 8), sharex='row')

traj1 = df1['trajectory'].dropna().iloc[0] if 'trajectory' in df1.columns else FILE_1
traj2 = df2['trajectory'].dropna().iloc[0] if 'trajectory' in df2.columns else FILE_2

plot_row(axes[0], df1, f'[1] {traj1}')
plot_row(axes[1], df2, f'[2] {traj2}')

for ax in axes[1]:
    ax.set_xlabel('Time (s)', fontsize=11)

fig.suptitle('Actual vs Reference — Log Comparison', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

## RMSE Comparison

Position RMSE = √( (x − x_ref)² + (y − y_ref)² + (z − z_ref)² )

> z values are already sign-flipped to ENU, so the RMSE is unaffected by the convention.

```python
def position_rmse(df: pd.DataFrame) -> float:
    err_sq = (df['x'] - df['x_ref'])**2 + (df['y'] - df['y_ref'])**2 + (df['z'] - df['z_ref'])**2
    return float(np.sqrt(err_sq.mean()))

def per_axis_rmse(df: pd.DataFrame):
    return {
        'x':   float(np.sqrt(((df['x']   - df['x_ref']  )**2).mean())),
        'y':   float(np.sqrt(((df['y']   - df['y_ref']  )**2).mean())),
        'z':   float(np.sqrt(((df['z']   - df['z_ref']  )**2).mean())),
        'yaw': float(np.sqrt(((df['yaw'] - df['yaw_ref'])**2).mean())),
    }

rmse1 = position_rmse(df1)
rmse2 = position_rmse(df2)
ax1   = per_axis_rmse(df1)
ax2   = per_axis_rmse(df2)

print(f"{'Metric':<28} {'File 1':>14} {'File 2':>14}")
print(f"{'File':<28} {FILE_1[:14]:>14} {FILE_2[:14]:>14}")
print("-" * 58)
print(f"{'Position RMSE (m)':<28} {rmse1:>14.4f} {rmse2:>14.4f}")
print(f"{'  RMSE x (m)':<28} {ax1['x']:>14.4f} {ax2['x']:>14.4f}")
print(f"{'  RMSE y (m)':<28} {ax1['y']:>14.4f} {ax2['y']:>14.4f}")
print(f"{'  RMSE z (m)':<28} {ax1['z']:>14.4f} {ax2['z']:>14.4f}")
print(f"{'  RMSE yaw (rad)':<28} {ax1['yaw']:>14.4f} {ax2['yaw']:>14.4f}")
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 4))

for ax, df, fname, rmse in [(axes[0], df1, FILE_1, rmse1), (axes[1], df2, FILE_2, rmse2)]:
    err = np.sqrt((df['x'] - df['x_ref'])**2 + (df['y'] - df['y_ref'])**2 + (df['z'] - df['z_ref'])**2)
    ax.plot(df['time'], err, lw=1.8)
    ax.axhline(rmse, color='red', ls='--', lw=1.5, label=f'RMSE = {rmse:.4f} m')
    ax.set_xlabel('Time (s)', fontsize=11)
    ax.set_ylabel('Position Error (m)', fontsize=11)
    ax.set_title(fname, fontsize=11)
    ax.legend(fontsize=10)

fig.suptitle('Position Error Over Time', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
metrics = ['x', 'y', 'z', 'yaw', 'position']
vals1   = [ax1['x'], ax1['y'], ax1['z'], ax1['yaw'], rmse1]
vals2   = [ax2['x'], ax2['y'], ax2['z'], ax2['yaw'], rmse2]

x = np.arange(len(metrics))
w = 0.35

fig, ax = plt.subplots(figsize=(10, 5))
bars1 = ax.bar(x - w/2, vals1, w, label=FILE_1, alpha=0.85)
bars2 = ax.bar(x + w/2, vals2, w, label=FILE_2, alpha=0.85)

ax.bar_label(bars1, fmt='%.4f', fontsize=8, padding=2)
ax.bar_label(bars2, fmt='%.4f', fontsize=8, padding=2)

ax.set_xticks(x)
ax.set_xticklabels([f'RMSE {m}' for m in metrics], fontsize=11)
ax.set_ylabel('RMSE', fontsize=11)
ax.set_title('Per-Axis and Position RMSE Comparison', fontsize=13, fontweight='bold')
ax.legend(fontsize=10)
plt.tight_layout()
plt.show()
```
