```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0, 15, 2000)
```

```python
# -------------------------------------------------
# SET 1 (original trajectories)
# -------------------------------------------------

# spiral (circle in XY)
R, T = 2.0, 7.5
x1 = R * np.sin(2*np.pi*t/T)
y1 = R * np.cos(2*np.pi*t/T)

# figure eight
R, T = 2.0, 10.0
x2 = R * np.sin(2*np.pi*t/T)
y2 = R * np.sin(4*np.pi*t/T)/2

# trefoil
R, T = 1.0, 15.0
s = 2*np.pi*t/T
x3 = R*(np.sin(s) + 2*np.sin(2*s))
y3 = R*(np.cos(s) - 2*np.cos(2*s))

# fig8 heading (same path as fig8)
R, T = 1.5, 15.0
s = 2*np.pi*t/T
x4 = R*np.sin(s)
y4 = R*np.sin(2*s)/2
import matplotlib.pyplot as plt
```

```python

# -------------------------------------------------
# SET 2 (new trajectories)
# -------------------------------------------------

# spiral (helix projected to XY)
R, T = 5.0, 7.5
x5 = R*np.sin(2*np.pi*t/T)
y5 = R*np.cos(2*np.pi*t/T)

# figure eight
R, T = 5.0, 10.0
x6 = R*np.sin(2*np.pi*t/T)
y6 = R*np.sin(4*np.pi*t/T)/2

# trefoil
R, T = 2.0, 15.0
s = 2*np.pi*t/T
x7 = R*(np.sin(s) + 2*np.sin(2*s))
y7 = R*(np.cos(s) - 2*np.cos(2*s))

# figure eight heading
R, T = 3.0, 15.0
s = 2*np.pi*t/T
x8 = R*np.sin(s)
y8 = R*np.sin(2*s)/2


# -------------------------------------------------
# Plot Set 1
# -------------------------------------------------

plt.figure(figsize=(8,8))
plt.plot(x1,y1,label="spiral")
plt.plot(x2,y2,label="figure_eight")
plt.plot(x3,y3,label="trefoil")
plt.plot(x4,y4,label="fig8_heading")
plt.title("Set 1 Trajectories")
plt.axis("equal")
plt.grid(True)
plt.legend()

# -------------------------------------------------
# Plot Set 2
# -------------------------------------------------

plt.figure(figsize=(8,8))
plt.plot(x5,y5,label="spiral")
plt.plot(x6,y6,label="figure_eight")
plt.plot(x7,y7,label="trefoil")
plt.plot(x8,y8,label="figure_eight_heading")
plt.title("Set 2 Trajectories")
plt.axis("equal")
plt.grid(True)
plt.legend()

plt.show()
```
